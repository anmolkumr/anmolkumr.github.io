#### Q1. Which books are most popular, based on ratings?


```python
#

import pandas as pd

books = pd.read_csv('books.csv')
bookrate = books[['title', 'average_rating']].groupby(['title']).mean()

topbooks = bookrate.sort_values('average_rating', ascending=False)
topbooks.head(5)
```

#### Q2. Effect of publisher on rating of Books

```python
import pandas as pd
import matplotlib.pyplot as plt

book = pd.read_csv('books.csv')
rating  = pd.read_csv('ratings.csv')

new_df = pd.merge(book, rating, on='book_id')
new_df[['title','authors']]

avg_books = new_df.groupby('book_id')['rating'].mean()
auth_avg = new_df.groupby('authors')['rating'].mean()

counts = new_df['authors'].value_counts()


plt.scatter(avg_books,book.groupby('book_id')['authors'].apply(lambda x: auth_avg[x.iloc[0]]))
plt.xlabel('Book Average Rating')
plt.ylabel('Average Author Rating')
plt.title('Effect of Authors on rating of Books')
plt.show()
```

#### Q3. Is there any Trend between Books count and Average rating


```python
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv('books.csv')

plot_data = file[['book_id','average_rating', 'books_count']]

plt.scatter(plot_data['books_count'], plot_data['average_rating'], alpha=0.6)
plt.title('Books Count vs. Average Rating')
plt.xlabel('Book Count')
plt.ylabel('Average Rating')
plt.show()
```

We can conclude that Highly rated books are less in count. 

Avaerage rated books are very popular or we can say that popular books are average rated.

####  Q5. Higher the number of ratings, higher the rating of the book

```python
import pandas as pd
import matplotlib.pyplot as plt

books_df = pd.read_csv('books.csv')

plt.scatter(books_df['ratings_count'], books_df['average_rating'], alpha=0.5)

plt.xlabel('Number of Ratings (in 10^6)')
plt.ylabel('Average Rating')
plt.title('Relationship between Number of Ratings and Average Rating')

# display the plot
plt.show()
```

This hypothesis is wrong that rating of the books are depending on number of ratings. As we can see in the plot that, no. of ratings are low but the Rating of book is high. Also, Books aroung 100000 ratings are below rating of 3.5
