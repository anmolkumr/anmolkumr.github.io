#### Q1. Hypothesis: If a player win Aces then He wins Game.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv('AusOpen-men-2013.csv')

sns.scatterplot(x='ACE.1', y='FNL1', data=df)
plt.title('Correlation between Aces Won and Winning')
plt.xlabel('Aces Won by Player 1')
plt.ylabel('Final Number of Games Won by Player 1')
plt.show()

# Use regression analysis to determine the correlation
X = df[['ACE.1']]
y = df['FNL1']
reg = LinearRegression()
reg.fit(X, y)
print(f"Correlation between Aces Won and Winning: {reg.coef_[0]:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv('AusOpen-men-2013.csv')


sns.scatterplot(x='ACE.2', y='FNL1', data=df)
plt.title('Correlation between Aces Won and Winning')
plt.xlabel('Aces Won by Player 2')
plt.ylabel('Final Number of Games Won by Player 2')
plt.show()

# Use regression analysis to determine the correlation
X = df[['ACE.2']]
y = df['FNL2']
reg = LinearRegression()
reg.fit(X, y)
print(f"Correlation between Aces Won and Winning: {reg.coef_[0]:.2f}")
```

The assumption that, If a player win Aces then He wins Game is not satisfactory because there is not strong correration between these two factors. But It can be noted that the correlation is non-negetive, thus we can conclude that the player who wins aces gets a little advantage.

#### Q3. Hypothesis: Higher the first serve winning percentage higher is the chance of winning the match.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('AusOpen-men-2013.csv')

# Calculate the winning percentage of players based on their first serve percentage
df['Winning'] = df.apply(lambda x: x['FNL1'] if x['FNL1'] > x['FNL2'] else 0, axis=1)
df['First Serve Winning Percentage'] = df.apply(lambda x: x['Winning'] / (x['FNL1'] + x['FNL2']) if x['FSP.1'] > x['FSP.2'] else 0, axis=1)

# Calculate the overall winning percentage of players with high first serve percentage
high_fsp = df[df['FSP.1'] > 60] # Assuming 60% or above is considered high
overall_winning_percentage = high_fsp['First Serve Winning Percentage'].mean()

print(f"Overall Winning Percentage of Players\n in AusOpen-MEN with High First Serve Percentage: {overall_winning_percentage:.2f}")
```

Q4. Hypothesis:Higher the Break Point Higher the Winnings



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("USOpen-men-2013.csv")

bp_created = df["BPC.1"]
result = df["FNL1"]

correlation = bp_created.corr(result)

# Plot a scatter plot to visualize the relationship between the two columns
plt.scatter(bp_created, result)
plt.title("Break Points Created vs. Result")
plt.xlabel("Break Points Created")
plt.ylabel("Result")
plt.show()
print(f"The correlation coefficient between Break Points Created and Result is {correlation:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("USOpen-men-2013.csv")

# Select the columns "Break Points Created" and "Result"
bp_created = df["BPC.2"]
result = df["FNL2"]

# Calculate the correlation coefficient between the two columns
correlation = bp_created.corr(result)

# Plot a scatter plot to visualize the relationship between the two columns
plt.scatter(bp_created, result)
plt.title("Break Points Created vs. Result")
plt.xlabel("Break Points Created")
plt.ylabel("Result")
plt.show()

print(f"The correlation coefficient between Break Points Created and Result is {correlation:.2f}")
```

This clearly shows that any player who creates break points wins the match because the correlation coeifficient is 0.77 and 0.71 in both cases which proves a strong positive correlation in our assumption.

Q7. What is the trend of first serve percentage for men and women across different rounds in tennis major tournaments?

```python
import pandas as pd
import matplotlib.pyplot as plt


men = pd.read_csv('Wimbledon-men-2013.csv')
women = pd.read_csv('Wimbledon-women-2013.csv')

# calculate the mean first serve percentage for each round
men_avg = men.groupby('Round')['FSP.1'].mean()
women_avg = women.groupby('Round')['FSP.1'].mean()

# plot the line graph
plt.plot(men_avg.index, men_avg.values, label='Men')
plt.plot(women_avg.index, women_avg.values, label='Women')
plt.xlabel('Round')
plt.ylabel('First Serve Percentage')
plt.title('Trend of First Serve Percentage for Men and Women Across Different Rounds')
plt.legend()
plt.show()
```

Q8. Is there any correlation between Double Faults and Unforced Errors for a particular player.

```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv('USOpen-Women-2013.csv')

# create scatter plot
plt.scatter(df['DBF.1'], df['UFE.1'])

# set x and y axis labels
plt.xlabel('Double Faults Committed')
plt.ylabel('Unforced Errors Committed')

# set title
plt.title('Double Faults vs Unforced Errors')

# show plot
plt.show()
```

```python
# calculate correlation
corr = df['DBF.1'].corr(df['UFE.1'])
print('Correlation between Double Faults and Unforced Errors: {:.2f}'.format(corr))
```
